```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

\begin{equation}
\begin{array}{ll}
\min_{x_1,x_2} & -x_1-x_2 \\
\text{s. a. }  & 50x_1+24x_2\leq 2400 \\
               & 30x_1+33x_2\leq 2100 \\
               & -x_1\leq -45 \\
               & -x_2\leq -5,
\end{array}
\end{equation}

```python
def x2_v1(x1):
    return (2400 - 50*x1)/24
def x2_v2(x1):
    return (2100 - 30*x1)/33
```

```python
x1 = np.linspace(43, 47)
```
