```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(2, 2*np.pi)
len(x)
print(x[:5])
```

```python
x = np.linspace(0, 2*np.pi, 100)
len(x)
```

```python
y = np.sin(x)
len(y)
```

```python
y[:5]
```

```python
plt.figure(figsize=(12,4))
plt.plot(x,y, "x-" , ms=7, label = "$\sin(x)$")
plt.plot(x,np.cos(x), label = "$\cos(x)$", lw = 5)
plt.xlabel("$x$", fontsize = 18)
plt.ylabel("$y$", fontsize = 18)
plt.title("Funciones")
plt.legend()
plt.grid()
plt.plot(x,0*x, lw = 1)
plt.show()
```

```python
import sympy as sym
sym.init_printing(use_latex='mathjax')
```

```python
sym.var('x')
sym.sin(x)*np.sin(3)
```

```python
sym.var("x", real = True)
```

```python
f= x**2; f
```

```python
df = sym.diff(f,x,1)
```

```python
# resolver f'(x)=0 y mostrar soluciones
xc = sym.solve(df, x)
xc
```

```python
# convertir f e una funciÃ³n que se pueda evaluar numÃ©ricamente 

f_num = sym.lambdify([x], f, 'numpy')
x_vec = np.linspace(-5, 5, 100)

# graficar
plt.plot(x_vec, f_num(x_vec), color = "cyan")
plt.grid()
plt.xlabel('$x$', fontsize = 18)
plt.ylabel('$x^2$', fontsize = 18)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return x**2
```

```python
def h(x,y):
    return np.sin(x)/np.tan(y)
```

```python
x = np.linspace(-5,5)
plt.plot(x,f(x),"--")
plt.plot(x,h(x,.1), "-.")
plt.plot(x,h(x,.4), "-")
plt.scatter(0,0,marker = "D",s=50,lw=50)
plt.show
```
